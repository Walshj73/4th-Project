# 1. Existing Architecture

To get a general idea of how Convolutional Neural Networks operate and what their overall structure looks like I will create 3 classical networks LeNet-5, Alexnet & VGG-16. The purpose of this is to get a quick overview of how these networks preform. These are not the actual classical networks these are simply tests built with the same structure these networks once used. 

## 1.1 LeNet-5

LeNet-5 is a rather small network. It has only convolutional layers along with 1 fully connected layer. The key difference between modern architecture and LeNet is that LeNet originally used tanh as it's activation function while modern network architectures use Relu as their activation function. The test below shows the differnce in using tanh with 3 convolutional layers.

---
### 1.1.1 Import the required libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, AveragePooling2D
from tensorflow.python.keras.optimizers import Adam, SGD, RMSprop, Adagrad, Adamax
import pickle

%matplotlib inline
```

### 1.1.2 Import the data

```python
X = pickle.load(open("X.pickle","rb"))
y = pickle.load(open("y.pickle","rb"))
```

### 1.1.3 Set a random seed

```python
np.random.seed(1)
```

### 1.1.4 Normalize the data

```python
X = X/255.0
```

### 1.1.5 Create the Model

```python
model = Sequential()
```

### 1.1.6 Add the Convolution Layers

```python
model.add(Conv2D(16, (3,3), input_shape = X.shape[1:]))
model.add(Activation("tanh"))
model.add(AveragePooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3)))
model.add(Activation("tanh"))
model.add(AveragePooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("tanh"))
model.add(AveragePooling2D(pool_size=(2,2)))
```

### 1.1.7 Adding the Dense Layers

```python
model.add(Flatten())
model.add(Dense(64))
model.add(Dropout(0.2))

model.add(Dense(1))
model.add(Activation("sigmoid"))
```

### 1.1.8 Compiling the Model

```python
model.compile(loss="binary_crossentropy",
             optimizer="Adam",
             metrics=['accuracy'])
```

### 1.1.9 Fit the Model

Mediocore results using this network the finall validation accuracy recorded was a mere 67% below in the models loss fails to converge.

```python
results = model.fit(X, y, batch_size=32, epochs=50, validation_split=0.2)
```

### 1.1.10 Plot the Models Loss
---
The plot below indicates that after 50 epochs the training loss did not converge but stayed on a steady path as the number of epochs increased. This example shows that small networks using the tanh activation function do not give good results. Later in this notebook you will see as I begin to use Relu the over loss of the network will converge.

```python
plt.plot(results.history["loss"])
plt.show()
```

### 1.1.11 Futher visualizations
---
Further investigation shows that as the epochs increase our training accuracy continues to increase while its loss steadily declines but does not converge. The validation accuracy began to increase but then began to fall while its validation loss began to increase. One observation for these results is that the model is infact overfitted as the training accuracy almost reaches a value of 100% while the final validation accuracy recorded was 67%.

```python
plt.plot(results.history['acc'])
plt.plot(results.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

## 1.2 AlexNet

AlexNet was a direct improvement upon LeNet. AlexNet instead of using tanh as its activation function it uses Relu. AlexNet also introduces the use of Dropout on its fully connected layers to tackle the issue of overfitting. AlexNet uses a Dropout value 0.5 on both fully connected layers. Without dropout AlexNet would substantially overfit, using dropout requires the use of more iterations to achieve convergence.

---

```python
model = Sequential()
```

### 1.2.3 Add the Dense Layers

```python
model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(128))
model.add(Activation("relu"))

model.add(Dense(256))
model.add(Activation("relu"))

model.add(Dense(1))
model.add(Activation("sigmoid"))
```

### 1.2.4 Compile the Model

# 12. Batch Size & Epoch Range Investigation

The batch size and epoch range are the final two hyper-parameters that will be used during grid search. My analysis of the batch size discovered that using a small batch size of ranges 32-64 worked the best when the epoch range was between 50-100. As the batch size increased the model would no longer converge and would need an increase in the epoch range to acheive convergence. This why I chose these ranges as the values I would use in my process of grid search which would scan the model and discover the best preforming model.

---
## 12.1 Batch Size
### 12.1.1 Create the Model

```python
model = Sequential()

model.add(Conv2D(16, (3,3), dilation_rate=2, strides=1, input_shape = X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

model.add(Conv2D(32, (3,3), dilation_rate=2, strides=1))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))
          
model.add(Conv2D(64, (3,3), dilation_rate=2, strides=1))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

model.add(Conv2D(128, (3,3), dilation_rate=2, strides=1))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation("relu"))

model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(32))
model.add(Activation("relu"))

model.add(Dense(1))
model.add(Activation("sigmoid"))

model.compile(loss="binary_crossentropy",
             optimizer="Adam",
             metrics=['accuracy'])
```
