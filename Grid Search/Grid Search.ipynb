## 1.1 ADAM
### 1.1.1 Import the Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, AveragePooling2D
from tensorflow.python.keras.optimizers import Adam, Adamax
import pickle
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

%matplotlib inline
```

### 1.1.2 Create the Model Function

```python
def create_model(kernel_initializer="normal", neurons=32, lr=0.01, decay=0.0):
    model = Sequential()
    
    model.add(Conv2D(16, (3,3), dilation_rate=2, strides=1, kernel_initializer=kernel_initializer, activation="relu", input_shape = X.shape[1:]))
    model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

    model.add(Conv2D(32, (3,3), dilation_rate=2, strides=1, activation="relu", kernel_initializer=kernel_initializer))
    model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

    model.add(Conv2D(64, (3,3), dilation_rate=2, strides=1, activation="relu", kernel_initializer=kernel_initializer))
    model.add(MaxPooling2D(pool_size=(2,2), padding="same"))

    model.add(Conv2D(128, (3,3), dilation_rate=2, strides=1, activation="relu", kernel_initializer=kernel_initializer))
    model.add(MaxPooling2D(pool_size=(2,2), padding="same"))
    
    model.add(Flatten())
    
    model.add(Dense(neurons, activation="relu", kernel_initializer=kernel_initializer))
    model.add(Dense(neurons, activation="relu", kernel_initializer=kernel_initializer))
    model.add(Dense(neurons, activation="relu", kernel_initializer=kernel_initializer))

    model.add(Dense(1, activation="sigmoid"))
    
    optimizer = Adam(lr=lr, decay=decay)
    
    model.compile(loss="binary_crossentropy",
             optimizer=optimizer,
             metrics=['accuracy'])
    return model
```

### 1.1.3 Set the Random Seed

```python
seed = 1
np.random.seed(seed)
```

### 1.1.4 Import the Data

```python
X = pickle.load(open("X.pickle","rb"))
y = pickle.load(open("y.pickle","rb"))
```

### 1.1.5 Normalize the Data

```python
X = X/255.0
```

### 1.1.6 Create the list of Hyper-Parameters

```python
kernel_initializer = ["normal", "glorot_uniform"]
neurons = [32, 64, 128]
lr = [0.01, 0.001]
decay = [0.0, 0.1]
batch_size = [32, 64]
epochs = [50, 100]
```

### 1.1.6 Create the Model

```python
model = KerasClassifier(build_fn=create_model, verbose=1)
```

### 1.1.7 Initialize the Parameters

```python
param_grid = dict(kernel_initializer=kernel_initializer, neurons=neurons, lr=lr, decay=decay, batch_size=batch_size, epochs=epochs)
```

### 1.1.8 Create the Grid

Due to Grid Search taking a substantial amount of time a value of cv=2 was used which sets the value of k-folds to be equal to number 2. This substantially decreases the training time and speeds up grid search. This was a cruicial step in this project in order to achieve an outcome from Grid Search. Before this Multiple runs of Grid Search where ran using a k-fold of 3 however as the process would not finish in time the number of k-folds had to be decreased.

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1, cv=2)
```

### 1.1.9 Fit the Model

```python
grid_result_adam = grid.fit(X, y)
```

### 1.1.10 Print the recorded Results

```python
print("Best: %f using %s" % (grid_result_adam.best_score_, grid_result_adam.best_params_))
means = grid_result_adam.cv_results_['mean_test_score']
stds = grid_result_adam.cv_results_['std_test_score']
params = grid_result_adam.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
